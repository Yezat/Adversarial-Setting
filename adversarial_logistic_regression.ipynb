## Aid Functions

```python
def epoch_adv(loader, model, delta):
    total_loss, total_err = 0.0, 0.0
    
    for X, y in loader:
        yp = model(
            (X - (2 * y.float()[:, None, None, None] - 1) * delta).view(X.shape[0], -1)
        )[:, 0]

        loss = nn.BCEWithLogitsLoss()(yp, y.float())
        
        total_err += ((yp > 0) * (y == 0) + (yp < 0) * (y == 1)).sum().item()
        total_loss += loss.item() * X.shape[0]
    
    return total_err / len(loader.dataset), total_loss / len(loader.dataset)
```

```python
def epoch_robust(loader, model, epsilon, opt=None, cov=None):
    total_loss, total_err = 0.0, 0.0

    if cov is None:
        cov = torch.eye(784)

        cov.requires_grad = False

    for X, y in loader:
        yp = model(X.view(X.shape[0], -1))[:, 0]

        scaling_factor = epsilon * (2 * y.float() - 1) / model.weight.norm(2)
        w_cov_w_scalar = (model.weight @ cov @ model.weight.T).item()
        yp = yp - scaling_factor * w_cov_w_scalar

        loss = nn.BCEWithLogitsLoss()(yp, y.float())

        if opt:
            opt.zero_grad()
            loss.backward()
            opt.step()

        total_err += ((yp > 0) * (y == 0) + (yp < 0) * (y == 1)).sum().item()
        total_loss += loss.item() * X.shape[0]

    return total_err / len(loader.dataset), total_loss / len(loader.dataset)
```

### making it into a funciton

```python
def train_in_adversarial(cov, epsilon_train, eps_attack, train_loader, test_loader, n_epochs=30):
    model = nn.Linear(784, 1, bias=False)
    opt = optim.SGD(model.parameters(), lr=1e-3)

    for i in range(n_epochs):
        _,_ = epoch_robust(train_loader, model, epsilon_train, opt, cov=cov)
        test_err, test_loss = epoch_robust(test_loader, model, epsilon_train, cov=cov)

    w = model.weight.detach().clone()
    delta = eps_attack * (w / w.norm(2)).view(28, 28)

    return epoch_adv(test_loader, model, delta)
```
